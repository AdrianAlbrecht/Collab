```python
import pandas as pd
import glob
import os
```

```python
# Ścieżka do katalogu zawierającego pliki CSV
sciezka_do_katalogu = ''
# Zidentyfikuj pliki CSV w danym katalogu
pliki_csv = glob.glob(sciezka_do_katalogu + '*.csv')
```

```python
ramka_danych = pd.concat((pd.read_csv(plik, sep=";", header=None) for plik in pliki_csv), ignore_index=True)
ramka_danych
```

```python
# Usuń kolumny zawierające same NaN
ramka_danych = ramka_danych.dropna(axis=1, how='all')
ramka_danych
```

```python
# Zidentyfikuj unikalne wartości w pierwszej kolumnie
unikalne_wartosci = ramka_danych.iloc[:, 0].unique()
# Stwórz słownik
slownik_wartosci = {klucz: [0]*8 for klucz in unikalne_wartosci}
# Stwórz słownik, gdzie klucze to unikalne wartości, a wartościami są ramki danych
ramki_danych_slownik = {wartosc: ramka_danych[ramka_danych.iloc[:, 0] == wartosc] for wartosc in unikalne_wartosci}
```

```python
check = []
```

```python
for klucz, ramka in ramki_danych_slownik.items():
  # Usuwamy NaN'y
  ramka = ramka.copy()
  ramka = ramka.dropna(axis=1, how='all')
  # Dzielimy na Single i Multi
  ramka['Single/Multi'] = 'Multi'
  indeksy_multi = ramka[(ramka.iloc[:, 1].diff() != 1) & (ramka.iloc[:, 1].shift(-1).diff() != 1)].index
  ramka.loc[indeksy_multi, 'Single/Multi'] = 'Single'
  # Określamy operację
  # insercja T - - - - -
  # delecja - T T T T T
  # substytucja C A A A A A
  # mixed A C T G C C C A
  ramka['Operation'] = 'Mixed'
  # Znajdź indeksy dla substytucji
  indeksy_substitution = ramka.loc[(ramka.iloc[:, 2] != '-') & (ramka.iloc[:, 3:-2].eq(ramka.iloc[:, 3], axis=0).all(axis=1))].index
  # Znajdź indeksy dla insercji
  indeksy_insertion = ramka.loc[(ramka.iloc[:, 2] != '-') & (ramka.iloc[:, 3:-2].eq("-", axis=0).all(axis=1))].index
  # Znajdź indeksy dla delecji
  indeksy_deletion = ramka.loc[(ramka.iloc[:, 2] == '-') & (ramka.iloc[:, 3:-2].eq(ramka.iloc[:, 3], axis=0).all(axis=1))].index
  # Ustaw wartość operacji dla odpowiednich wierszy w kolumnie "Operation"
  ramka.loc[indeksy_substitution, 'Operation'] = 'Substitution'
  ramka.loc[indeksy_insertion, 'Operation'] = 'Insertion'
  ramka.loc[indeksy_deletion, 'Operation'] = 'Deletion'
  check_duplicate = ramka.copy()
  # Indeksy do usuniecia (Multi ma być liczone jako jeden)
  indeksy_do_usuniecia = ramka[(ramka['Single/Multi'] == 'Multi') & (ramka['Single/Multi'].shift(-1) == 'Multi')\
                                      & (ramka.iloc[:, 1] + 1 == ramka.iloc[:, 1].shift(-1))].index
  # Usuń niepotrzebne wiersze multi
  ramka = ramka.drop(index=indeksy_do_usuniecia)
  #Liczymy wartości
  slownik_wartosci[klucz] = [\
      ((ramka.iloc[:, -1] == 'Substitution') & (ramka.iloc[:, -2] == 'Single')).sum(),\
      ((ramka.iloc[:, -1] == 'Deletion') & (ramka.iloc[:, -2] == 'Single')).sum(),\
      ((ramka.iloc[:, -1] == 'Insertion') & (ramka.iloc[:, -2] == 'Single')).sum(),\
      ((ramka.iloc[:, -1] == 'Mixed') & (ramka.iloc[:, -2] == 'Single')).sum(),\
      ((ramka.iloc[:, -1] == 'Substitution') & (ramka.iloc[:, -2] == 'Multi')).sum(),\
      ((ramka.iloc[:, -1] == 'Deletion') & (ramka.iloc[:, -2] == 'Multi')).sum(),\
      ((ramka.iloc[:, -1] == 'Insertion') & (ramka.iloc[:, -2] == 'Multi')).sum(),\
      ((ramka.iloc[:, -1] == 'Mixed') & (ramka.iloc[:, -2] == 'Multi')).sum()\
  ]
  check = ramka.copy()
```

```python
# Przekształć słownik na ramkę danych
dataset = pd.DataFrame(list(slownik_wartosci.items()), columns=['Species', 'Values'])
# Rozbij listę na pojedyncze wartości i umieść jako nowe kolumny
dataset[['Single Substitution', 'Single Deletion', 'Single Insertion', 'Single Mixed', \
          'Multi Substitution', 'Multi Deletion', 'Multi Insertion', 'Multi Mixed']] = pd.DataFrame(dataset['Values'].tolist(), index=dataset.index)
# Usuń niepotrzebną kolumnę 'Wartosci'
dataset = dataset.drop('Values', axis=1)
dataset
```

```python
if not os.path.exists("wynik"):
    # Jeżeli folder nie istnieje, utwórz go
    os.makedirs("wynik")
dataset.to_csv('wynik/MDCS_without_length.csv', sep=";", index=False)
```
