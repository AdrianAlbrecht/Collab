# Wszystko w porządku. Możemy przejść do reformatowania danych. W kolumnach "ca" oraz "tha" możemy zamienić format danych z float na int, ponieważ żadna z danych nie ma wartości po przecinku- ułatwi to dokładność operacji numerycznych na tych danych. Nastpęnie stworzymy kopię danych, ponieważ w dalszych krokach (analiza danych) łatwiej jest się poruszać po danych zdyskretyzowanych, a w kopii wykonamy zamianę 0 na mężczyznę(M), a 1 na kobietę (K) w kolumnie "sex" kierując się, że 1 to obecność allelu Y typowego dla płci żeńskiej.

```python
df["ca"] = df['ca'].astype(int)
df['ca'].values
```

```python
df["thal"] = df['thal'].astype(int)
df['thal'].values
```

# Stworzymy kopię zdyskretyzowanych danych:

```python
df_after_discretization = df.copy()
```

# A teraz zamienimy wartości w kolumnie płci:

```python
df['sex'] = df['sex'].astype(str)
df['sex'] = df['sex'].map({'0': 'M', '1': 'K'})
df['sex'].values
```

# Możemy jeszcze odwrócicć proces sykretyzacji dla decyzji (kolumna "num"), gdzie 0 to "brak zawału", a reszta liczb to kolejno "Zawał typu X":

```python
# Definiowanie funkcji do mapowania
def map_num(num):
    if num == '0':
        return "brak"
    else:
        return f"Zawał typu {num}"
```

```python
df['num'] = df['num'].astype(str)
df['num'] = df['num'].map(map_num)
df['num'].values
```

# Możemy wcześniej sprawdzane warunki 0 lub 1 (oprócz kolumny "sex") zamienić na True/False.

```python
df['exang'] = df['exang'].astype(bool)
df['exang'].values
```

```python
df['fbs'] = df['fbs'].astype(bool)
df['fbs'].values
```
